```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_excel('Scoring02.xlsx')
df
```

```python
df.loc[df['Пол'] == 'Муж' , 'gender'] = 1
df.loc[df['Пол'] == 'Жен' , 'gender'] = 0
df = df.drop(['Пол'] , axis = 1)
df
```

```python
df = df.drop(["Возраст"], axis = 1)
df = df.drop(["Год"], axis = 1)
df.info()
```

```python
df['Образование'].value_counts()
```

```python
df.loc[df['Образование'] == 'Среднее образование' , 'education'] = 1
df.loc[df['Образование'] == 'Высшее образование' , 'education'] = 2
df.loc[df['Образование'] == 'Сред.спец.образ-ние' , 'education'] = 3
df.loc[df['Образование'] == 'Непол Сред.образ' , 'education'] = 4
df.loc[df['Образование'] == 'Начал образование' , 'education'] = 5
df.loc[df['Образование'] == 'Аспирантура' , 'education'] = 6
df = df.drop(['Образование'] , axis = 1)
df.info()
```

```python
df.loc[df['Семейное положение'] == 'Married' , 'family'] = 1
df.loc[df['Семейное положение'] == 'Single' , 'family'] = 2
df.loc[df['Семейное положение'] == 'Divorced' , 'family'] = 3
df.loc[df['Семейное положение'] == 'Widow/Widower' , 'family'] = 4
df = df.drop(['Семейное положение'] , axis = 1)
df
```

```python
df.loc[df['Повторный клиент'] == 'Старый клиент' , 'customer'] = 1
df.loc[df['Повторный клиент'] == 'Новый клиент' , 'customer'] = 2
df = df.drop(['Повторный клиент'] , axis = 1)
```

```python

df.loc[df['Тип бизнеса'] == '1. Карзи истеъмоли/Потребительский кредит' , 'tp'] = 1
df.loc[df['Тип бизнеса'] == '3. Хизматрасони/Услуги' , 'tp'] = 2
df.loc[df['Тип бизнеса'] == '4. Савдо / Торговля' , 'tp'] = 3
df.loc[df['Тип бизнеса'] == '2. Истехсолот/Производство' , 'tp'] = 4
df.loc[df['Тип бизнеса'] == '6. Хочагии кишлок / Сельское хозяйство' , 'tp'] = 5
df = df.drop(['Тип бизнеса'] , axis = 1)
df
```

```python
df.loc[df['Длительность самой долгой единовременной просрочки в течение цикла'] <= 20 , 'result'] = 1
df.loc[df['Длительность самой долгой единовременной просрочки в течение цикла'] >=21 , 'result'] = 2
df = df.drop(['Длительность самой долгой единовременной просрочки в течение цикла'] , axis = 1)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('result', axis=1), df['result'], test_size=0.20, random_state= 100)
clf = RandomForestClassifier(n_estimators=100, max_depth=7, random_state= 100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = np.mean(y_pred == y_test)
accuracy
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, y_pred)
auc
```
